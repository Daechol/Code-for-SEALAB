```python
import os
import arrow
import datetime
import time
```

```python
file_path = 'directory'
```

```python
file_names = os.listdir(file_path)
```

```python
file_names_nc = [file for file in file_names if file.endswith(".nc")]
```

```python
file_names_nc
```

```python
for name in file_names_nc:
    
    src = os.path.join(file_path, name)

    arr = name.split('%')
    dateFormatter = '%Y-%m-%d+%H'
    arr1 = datetime.datetime.strptime(arr[0], dateFormatter)

    timedelta_ = datetime.datetime.now() - datetime.datetime.utcnow()
    utc_kst_hours = timedelta_.seconds // 3600

    timezon_kst = datetime.timezone(datetime.timedelta(hours=utc_kst_hours))
    dt_timezone = arr1.replace(tzinfo=timezon_kst)

    dt_timezone = dt_timezone.astimezone(datetime.timezone.utc)

    dst = f"GK2_GOCI2_L1B_{dt_timezone.strftime('%Y%m%d_%H')}1530_LA_S007.nc"
    dst = os.path.join(file_path, dst)
    
    os.rename(src, dst)
```
